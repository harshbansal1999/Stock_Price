```python
def step_gradient(points,Y_train, learning_rate, m):
    m_slope = np.zeros(15)
    new_m = np.zeros(15)
    M=len(points)
    s=0
    for i in range(M):
        s=0
        y=Y_train[i]
        for j in range(15):
            s+=m[j]*points[i][j]
        for k in range(15):
            m_slope[k] += (-2/M)*(y-s)*points[i,k]
    for i in range(15):
        new_m[i] = m[i]-learning_rate*m_slope[i]
    return new_m
```

```python
def cost(points,Y_trains,  m):
    total_cost = 0
    M = len(points)
    s=0
    for i in range(M):
        s=0
        y=Y_trains[i]
        for j in range(15):
            s+=m[j]*points[i,j]
        total_cost+=(1/M)*((y-s)**2)
        s=0
    return total_cost
            
```

```python
def gd(points, Y_train, learning_rate, num_iterations):
    m=np.zeros(15)
    for i in range(num_iterations):
        m=step_gradient(points,Y_train,learning_rate,m)
        print( " Cost: ", cost(points, Y_train,m))
    return m
```
